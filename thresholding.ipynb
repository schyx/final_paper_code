### Soft vs. hard thresholding illustrated

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cutoff = 1
xs = np.linspace(-3, 3, 1000)
```

```python
def f_hard(x, cutoff):
    if abs(x) > cutoff:
        return x
    return 0

def f_soft(x, cutoff):
    sign = np.sign(x)
    return max(abs(x)-cutoff, 0)*sign
```

```python
y_hard = [f_hard(x, cutoff) for x in xs]
y_soft = [f_soft(x, cutoff) for x in xs]
```

```python
plt.plot(xs, y_hard, label="hard cutoff at 1")
plt.plot(xs, y_soft, label="soft cutoff at 1")
plt.legend()
plt.grid(True)
plt.savefig('bryant_figures/activation.png', dpi=150)
```
